```python
# Importing File

import password
from datetime import date
```

```python
# Database:

books = {
    
    'class_1' : {
        
        'Hindi' : {
            
            'Copies' : ['h101','h102','h103','h104','h105','h108'],
            'Issued' : ['h106','h107']
        },
        
        'English' : {
            
            'Copies' : ['e101','e102','e103','e104','e105','e108'],
            'Issued' : ['e106','e107']
        }
    },
    
    'class_2' : {
        
        'Maths' : {
            
            'Copies' : ['m201','m202','m203','m204','m205','m208'],
            'Issued' : ['m206','m207']
        },
        
        'English' : {
            
            'Copies' : ['e201','e202','e203','e204','e205','e208'],
            'Issued' : ['e206','e207']
        }
    }
}

Acc = {
    
    'Admin_1001' : 'AdminLib1234@' ,
    'C101' : '1234'
}
acc_ = {'adm' : ['Admin_1001'],
       'stu':['C101']
       }
std = {
    'C101' : {
        
        'name'     : 'Ashwin Sharma',
        'class'    : 1,
        'book'     : {'Hindi' : { 'b_id' : 'h106', 'Date' : date(2024,4,2)} ,'English' : { 'b_id' : 'e107', 'Date' : date(2024,4,2)}},
        'fine'     : 0

    },
    
    'C201' : {
        'name'     : 'Namiya Bano' ,
        'class'    : 2,
        'book'     : {'Maths' : { 'b_id' : 'm206', 'Date' : date(2024,4,2)} ,'English' : { 'b_id' : 'e207', 'Date' : date(2024,4,2)}},
        'fine'     : 0
      
      
      }
}
```

```python
def fine_updater():
    for i in std.keys():
        t = 0
        for j in std[i]['book'].keys(): 
            b = str(date.today() - std[i]['book'][j]['Date'])
            c  = int(b.split()[0]) - 7
            c  = int(c) * 2
            if c > 0:
                t +=c
        std[i]['fine'] = t
        
```

### Sign up and login is below

```python
def sign_up():
    a = input('Please Enter Your School ID: ').title() 
    b = password.pass_checker()
    
    Acc.update({a:b})
    acc_['stu'].append(a)
    std.update({a: {'name' : input('Enter Your Name: '), 'class' : int(input('Enter your class')), 'book': {}, 'fine': 0}})
    stud(a)
    
def login():
    a = input('Enter Your ID: ').title()
    if a in Acc.keys():
        b = input('Enter Your Password: ')
        if b == Acc[a]:
            if a in acc_['adm']:
                print(' Welcome Admin '.center(50))
                admin()
            else:
                print(f'Welcome Student{a}'.center(50))
                a  = a.upper()
                stud(a)
        else:
            print('Wrong Password')
            login()
    else:
        print('Please Create An Account Before Accessing The Library OR Input Your ID Again If You Have Signed Up Already')
        print('Exiting The Library')
```

### Admin

```python
def admin():
    a = int(input('''
    Enter 1 to check total books
    Enter 2 to Update fine 
    Enter 3 to show student details who had issued a book
    Enter 4 to Exit The Library
    '''))
    
    if a == 1:
        for i in books.keys():
            print(i.center(70).upper())
            print('-'*75)
            print('|'+'SUBJECT'.center(15) + '|'+'Available Copies'.center(20) + '|'+'Issued Copies'.center(20) + '|'+'Total'.center(15) + '|')
            print('-'*75)
            for j in books[i].keys():
                print('|'+j.center(15) + '|'+ str(len(books[i][j]['Copies'])).center(20) + '|'+str(len(books[i][j]['Issued'])).center(20) + '|'+str(len(books[i][j]['Copies']) + len(books[i][j]['Issued'])).center(15) + '|')
                print('-'*75)
            print()
    elif a == 2:
        fine_updater()
        for i in std.keys():
            a = std[i]['fine']
            print(i,'==>', a)
    elif a == 3:
        print('+' + '-'*7  +'+' +'-'*20 + '+' +'-'*25 +'-'*5  + '+' +'-'*10 + '+'  )
        print('|'+'Class'.center(7)+'|' +'Student Name'.center(20)+'|' +'Book ID'.center(30)+'|'+'Fine'.center(10)+'|')
        print('+' + '-'*7  +'+' +'-'*20 + '+' +'-'*25 +'-'*5  + '+' +'-'*10 + '+'  )
        for i in std.keys():
            print('|'+ str(std[i]['class']).center(7)+'|'+ std[i]['name'].center(20) +'|'+ ', '.join(list(std[i]['book'].keys())).center(30) +'|'+str(std[i]['fine']).center(10)+'|')
            print('+' + '-'*7  +'+' +'-'*20 + '+' +'-'*25 +'-'*5  + '+' +'-'*10 + '+'  )
    elif a == 4:
        print('Exiting The Library')
    else:
        print('Wrong Input')
        admin()
```

# Student

> details check

> submit a book

> Issue a book

> close

```python
def stud(ID):
    a = int(input('''
    PRESS 1 TO ISSUE A BOOK
    PRESS 2 TO SUBMIT A BOOK
    PRESS 3 TO CHECK DETAILS
    PRESS 4 TO EXIT
    '''))
    if a == 1:
        # Shows the books available to issue          ########
        a = 'class_'+str(std[ID]['class'])
        b = []
        for i in books[a].keys():
            if books[a][i]['Copies'] != []:
                b.append(i)
        print("Available books are ", b)

        # Issue book code below:                   ########
        bk = input('Enter The Name Of Book You Want To Issue: ').title()
        b = books[a][bk]['Copies'].pop()
        books[a][bk]['Issued'].append(b)
        std[ID]['book'].update({b:{'b_name' : bk, 'Date' : date.today()}})
        print('Your book has been issued')
        
        
    elif a == 2:
        a = std[ID]['class']
        a = 'class_'+str(a)
        b = list(input('Enter the Books ID with space between them in case of multiple books to submit: ').strip().split(' '))
        for i in b:
            f = std[ID]['book'][i]['b_id']
            books[a][i]['Issued'].remove(f)
            books[a][i]['Copies'].append(f)
            std[ID]['book'].pop(i)
        else:
            print('Your Book has been Submitted')
    
    elif a == 3:
        fine_updater()
        print('+' + '-'*15 + '+' + '-'*7 + '+' + '-'*21 + '+' + '-'*6 + '+')
        print('|'+ 'Name'.center(15) +'|'+ 'Class'.center(7) +'|' + 'Issued Books ID'.center(21) +'|' +'Fine'.center(6)+'|')
        print('+' + '-'*15 + '+' + '-'*7 + '+' + '-'*21 + '+' + '-'*6 + '+')
        print('|'+ std[ID]['name'].center(15) +'|'+ str(std[ID]['class']).center(7) +'|' + ', '.join(list(std[ID]['book'].keys())).center(21) +'|' + str(std[ID]['fine']).center(6)+'|')
        print('+' + '-'*15 + '+' + '-'*7 + '+' + '-'*21 + '+' + '-'*6 + '+')
        
    elif a == 4:
        print('Exiting The Library')
    else:
        print('Wrong choice')
        stud(ID)
```
